```
rawdata = pd.read_csv('https://api.covid19india.org/csv/latest/raw_data.csv')
```

```
rawdata['Date Announced'] = pd.to_datetime(rawdata['Date Announced'], format='%d/%m/%Y')
```

```
rawdata.sort_values(['Date Announced']).head(200).groupby('Detected City')[['Patient Number']].count().sort_values(by ='Patient Number', ascending=False)
```

```
rawdata.sort_values(['Date Announced']).head(200).groupby('Detected State')[['Patient Number']].count().sort_values(by ='Patient Number', ascending=False)
```

```
rawdata.groupby(['Contracted from which Patient (Suspected)'])['Patient Number'].count().sort_values(ascending=False)[:10].to_frame()
```
